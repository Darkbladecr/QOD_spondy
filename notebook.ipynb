**QOD Spondylolithesis Study**

`XX` is related to time:
- BL = baseline
- 3m
- 12m
- 24m
- 36m
- 60m
- latest

Variable | Description
--- | ---
`id` | unique patient identifier
`site` | hospital site
`site_id` | hospital site id
`date_sx` | month and year of index surgery
`date_sx_A` | ???
`age` | years
`sex` | 1 = male, 2 = female
`principal_spondy_dx` | 1 = spondylolisthesis (some patients had a primary dx of lumbar stenosis and later found to have radiological spondy dx)
`grade_listhesis` | [Meyerding spondylolisthesis grade](https://radiopaedia.org/articles/spondylolisthesis-grading-system)
`height` | centimeter height
`weight` | kilogram weight
`insurance` | 1 = uninsured<br />2 = Medicare<br />3 = Medicaid<br />4 = VA/Government<br />5 = Private
`prev_sx` | 1 = yes (previous surgery)
`smoker` | 0 = no<br />1 = yes, current smoker<br />2 = unknown
`diabetes` | 1 = yes
`cad` | 1 = yes
`anxiety` | 1 = yes
`depression` | 1 = yes
`osteoporosis` | 1 = yes
`main_symptom` | 1 = back dominant<br />2 = leg dominant<br />3 = back-leg
`motor_deficit` | 1 = yes
`ambulation` | 1 = independently ambulatory<br />2 = ambulatory with an assistive device<br />3 = wheelchair bound
`symptom_duration` | 1 = <3 months<br />2 = >3 months<br />3 = unknown
`bmi` | kg/m<sup>2</sup>
`ethnicity` | 1 = hispanic<br />2 = non-hispanic<br />3 = prefer not to answer
`education` | 1 = < High school<br />2 = High School Diploma or GED<br />3 = Two year college degree<br />4 = Four year college degree<br />5 = Post-College<br />6 = Prefer not to answer
`workers_comp` | 0 = no<br />1 = yes<br />3 = unknown
`employment` | 1 = employed and currently working<br />2 = Employed but not working (short-term disability or on leave)<br />3 = Unemployed<br />4 = Attending school
`unemployed_status` | 1 = On disability<br />2 = Retired<br />3 = Homemaker<br />4 = None of the above
`asa_grade` | 1 = healthy<br />2 = mild systemic disease<br />3 = severe systemic disease<br />4 = severe systemic disease that is a constant threat to life<br />5 = moribund patient who is not expected to survive without the operation<br />6 = brain-dead patient
`surg_approach` | 1 = posterior only<br />2 = anterior only<br />3 = lateral only<br />4 = two-stage procedure
`laminectomy_performed` | 1 = yes
`laminectomy_levels` | number of intervertebral segments operated on (e.g. L1-2 laminectomies = 1 level, L1 laminectomy = 1 level)
`athrodesis_performed` | 1 = yes
`athrodesis_levels` | _same as laminectomy levels_
`interbody_graft` | 1 = yes
`MIS_decompression` | 1 = yes
`MIS_percurtaneous_pedicle_screws` | 1 = yes
`MIS_pedicle_screws` | 1 = yes
`cortical_screws` | 1 = yes
`MIS_interbody` | 1 = yes
`blood_loss` | milliliters
`length_of_surgery` | minutes
`los` | length of hospitalization (days)
`place_discharged_to` | 1 = Home routine<br />2 = Home with home heatlhcare service<br />3 = Post-acute or non-acute care setting<br />4 = Transferred to another acute care facility<br />5 = Against Medical Advice or AMA<br />6 = Died in hospital<br />7 = Discharged alive, destination unknown
`place_discharged_to_facility` | **if placed_discharged_to = 3**<br />1 = skilled nursing facility<br />2 = intermediate/rehabilitation care facility<br />3 = another type of facility
`XX_pt_satisfaction_index` | 1 = surgery met my expectations<br />2 = I did not improve as much as I had hoped, but I would undergo the same surgery for the same outcome<br />3 = I did not improve as much as I had hoped, and I would not undergo the same surgery for the same outcome<br />4 = I am the same or worse than before surgery
`XX_return_to_work` | 0 = no<br />1 = yes<br />2 = not applicable
`XX_NRSBP` | values 1-10 (most severe = 10) [numeric rating scale for low back pain]
`XX_NRSLP` | values 1-10 (most severe = 10) [numeric rating scale for leg pain]
`XX_ODI` | values 0 - 100 (most disabled = 100) [Oswestry Disability Index]
`XX_EQ5D` | values 0 - 1.00 (best quality of life = 1.00) [Health related QOL score]
`mortality_30_day` | 1 = yes
`readmit_(30day, 3month)` | 1 = yes
`readmit_(30day, 3month)_reason` | free text
`readmit_(30day, 3month)_related` | free text
`overall_readmit_3month` | free text

`reop_revision_2-3year_related` | free text
