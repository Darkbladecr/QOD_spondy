```python
timeline = [
  "BL", 
  "3m", 
  "12m", 
  "24m", 
  "36m", 
  "60m", 
  "latest"
]

score_keys = [
  "NRSBP",
  "NRSLP",
  "EQ5D",
  "ODI"
]

score_keys_no_latest = [
  "ODI_pain_intensity",
  "ODI_persol_care",
  "ODI_lifting",
  "ODI_walking",
  "ODI_sitting",
  "ODI_standing",
  "ODI_sleeping",
  "ODI_sex_life",
  "ODI_social_life",
  "ODI_travelling",
  "mobility",
  "self_care",
  "usual_activities",
  "pain",
  "anxiety_depression"
]

score_keys_no_BL = [
  "pt_satisfaction",
  "return_to_work",
]

def append_key(df, arr, t, k):
  tmp = "{0}_{1}".format(t, k)
  if tmp in df.columns:
    arr.append(tmp)
  else:
    print("key doesn't exist:", tmp)

score_cols = []
for t in timeline:
  if t != "BL":
    for k in score_keys_no_BL:
      append_key(df, score_cols, t, k)
  if t != "latest":
    for k in score_keys_no_latest:
      append_key(df, score_cols, t, k)
  for k in score_keys:
    append_key(df, score_cols, t, k)

df_scores = df[score_cols]
df_scores.head()
```

```python
from missforest.miss_forest import MissForest
from os.path import exists

pkl_file = './missForestDF.pkl'

if exists(pkl_file):
  dff = pd.read_pickle(pkl_file)
else:
  mf = MissForest()
  df_scores_filled = mf.fit_transform(df_scores.copy())
  dff = df.copy()
  dff[score_cols] = df_scores_filled
  dff[score_cols] = dff[score_cols].round().astype('Int64')
  dff.to_pickle(pkl_file)

print('Any null values:', dff[score_cols].isnull().values.any())
display(dff[score_cols].head())
```

```python
def value_to_pct(col):
  values = col.value_counts().astype(str)
  pct = col.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
  return (values + ' (' + pct + ')')

df_sleep_all_pct = pd.DataFrame({
  'Baseline': value_to_pct(df_sleep['BL_ODI_sleeping']),
  '3 mo': value_to_pct(df_sleep['3m_ODI_sleeping']),
  '12 mo': value_to_pct(df_sleep['12m_ODI_sleeping']),
  '24 mo': value_to_pct(df_sleep['24m_ODI_sleeping']),
}).fillna('0 (0%)')
df_sleep_impaired_pct = pd.DataFrame({
  'Baseline': value_to_pct(df_sleep[bool_impaired_BL]['BL_ODI_sleeping']),
  '3 mo': value_to_pct(df_sleep[bool_impaired_BL]['3m_ODI_sleeping']),
  '12 mo': value_to_pct(df_sleep[bool_impaired_BL]['12m_ODI_sleeping']),
  '24 mo': value_to_pct(df_sleep[bool_impaired_BL]['24m_ODI_sleeping']),
}).fillna('0 (0%)')
df_sleep_normal_pct = pd.DataFrame({
  'Baseline': value_to_pct(df_sleep[bool_normal_BL]['BL_ODI_sleeping']),
  '3 mo': value_to_pct(df_sleep[bool_normal_BL]['3m_ODI_sleeping']),
  '12 mo': value_to_pct(df_sleep[bool_normal_BL]['12m_ODI_sleeping']),
  '24 mo': value_to_pct(df_sleep[bool_normal_BL]['24m_ODI_sleeping']),
}).fillna('0 (0%)')

df_sleep_all_pct.index.name = 'ODI Item 7 Score'
df_sleep_impaired_pct.index.name = 'ODI Item 7 Score'
df_sleep_normal_pct.index.name = 'ODI Item 7 Score'
```

```python
from scipy.stats import chi2_contingency
```

```python
chisqt = pd.crosstab(df_sleep['3m_ODI_sleeping'] == 0, df_sleep['3m_ODI_sleeping'])
value = np.array([chisqt.iloc[0].values, chisqt.iloc[1].values])
chi2, p, dof, _ = chi2_contingency(value)
if p < 0.001:
  p = "<0.001"
else:
  p = '{:.3f}'.format(p)
print('p =', p)
print('chi2 =', chi2)
print('DOF =', dof)
```
